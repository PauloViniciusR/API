```python
import requests
```

```python
URL_DDD = "https://brasilapi.com.br/api/ddd/v1/"

resposta_ddd_11 = requests.get("".join([URL_DDD, "11"]))

resposta_ddd_11.status_code
```

```python
resposta_ddd_11.headers
```

```python
resposta_ddd_11.json()
```

```python
from pprint import pprint

pprint(resposta_ddd_11.json())
```

```python
# criando dataframe para armazenar os dados do DDD 21
import pandas as pd 

ddd_11 = pd.DataFrame(resposta_ddd_11.json()).assign(ddd=11)

ddd_11
```

**ATENÇÃO**: Sempre verifique a documentação da API para verificar se há limites de requisições por minuto ou de loops.

```python
# verificando o status code de buscas por DDD entre 20 e 30
for ddd in range (11, 31):
    resposta = requests.get("".join([URL_DDD, str(ddd)]))
    print(f"DDD: {ddd} - Status code: {resposta.status_code}")
```

```python
# para os status code 200, vamos armazenar os dados em um dataframe
df_ddd = pd.DataFrame()

for ddd in range(11, 30):
    resposta = requests.get("".join([URL_DDD, str(ddd)]))
    if resposta.status_code == 200:
        df_ddd = pd.concat([df_ddd, pd.DataFrame(resposta.json()).assign(ddd=ddd)])

df_ddd

# verificando o status code e adicionando ao dataframe com o método concat
```

```python
df_ddd = df_ddd.reset_index(drop=True)

df_ddd
```

```python
df_ddd.info()
```

## Dados de Municípios da BrasilAPI

```python
URL_MUNICIPIOS_SP = "https://brasilapi.com.br/api/ibge/municipios/v1/SP?providers=dados-abertos-br,gov,wikipedia"

resposta_municipios_sp = requests.get(URL_MUNICIPIOS_SP)

resposta_municipios_sp.status_code
```

```python
resposta_municipios_sp.json()
```

```python
# convertendo para dataframe

df_sp = pd.DataFrame(resposta_municipios_sp.json())

df_sp
```

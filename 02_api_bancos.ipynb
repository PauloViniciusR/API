## Passos de consumo de uma API


1. **Estudo da Documentação:**
   - Antes de tudo, familiarize-se com a documentação da API. Ela fornecerá informações sobre como a API funciona, seus endpoints, parâmetros disponíveis, formatos de resposta e muito mais.
   
2. **Entendimento dos Status Codes:**
   - Compreenda os status codes HTTP para interpretar as respostas da API. Isso ajudará a diagnosticar problemas e entender o status das suas requisições.

3. **Registro e Autenticação:**
   - Muitas APIs exigem que você se registre para obter uma chave de API ou token. Esse token é usado para autenticar suas requisições e, muitas vezes, para rastrear e limitar seu uso.

4. **Configuração do Ambiente:**
   - Configure seu ambiente de desenvolvimento. Isso pode envolver a instalação de bibliotecas ou ferramentas específicas, como `requests` em Python.

5. **Realização de Requisições Teste:**
   - Comece fazendo requisições simples para testar a conexão e entender a estrutura das respostas. Isso pode envolver a obtenção de um único item ou listar vários itens.

6. **Tratamento de Parâmetros e Cabeçalhos:**
   - Aprenda a enviar parâmetros (por exemplo, para filtrar resultados) e a definir cabeçalhos adequados (como tokens de autenticação ou informações de tipo de conteúdo).

7. **Paginação e Limites:**
   - Entenda como a API lida com grandes conjuntos de dados. Muitas APIs paginam os resultados ou impõem limites de quantos itens você pode solicitar de uma vez.

8. **Tratamento de Respostas e Erros:**
   - Analise as respostas, extraia os dados necessários e implemente tratamentos de erro adequados para lidar com possíveis falhas ou limitações.

9. **Integração com Aplicações:**
   - Integre os dados ou funcionalidades da API em sua aplicação, seja ela um site, aplicativo móvel ou sistema backend.

10. **Otimizações (Cache, Rate Limiting, etc.):**
   - Implemente otimizações, como caching de respostas frequentes, para melhorar o desempenho e garantir que você não exceda os limites da API.

11. **Monitoramento e Manutenção:**
   - Monitore o uso da API, verifique regularmente a saúde das requisições e esteja atento a quaisquer mudanças ou atualizações na API.




## Status Codes



### **O que são status codes?**

Status codes, ou códigos de status, são códigos de três dígitos que os servidores HTTP enviam para indicar o resultado de uma requisição feita por um cliente. Eles informam ao cliente se a requisição foi bem-sucedida, se houve algum erro, ou se é necessário realizar alguma ação adicional. Esses códigos são padronizados e fazem parte do protocolo HTTP, permitindo que clientes e servidores de diferentes implementações se comuniquem de forma eficaz.

### **Principais status codes e seus significados**

- **2xx: Sucesso**
  - **200 OK:** A requisição foi bem-sucedida. O resultado depende do método HTTP utilizado.
  - **201 Created:** A requisição foi bem-sucedida e um novo recurso foi criado como resultado.
  - **204 No Content:** A requisição foi bem-sucedida, mas não há conteúdo para enviar na resposta.

- **3xx: Redirecionamento**
  - **301 Moved Permanently:** O recurso solicitado foi movido permanentemente para uma nova URL.
  - **302 Found (ou Moved Temporarily):** O recurso solicitado foi movido temporariamente para uma nova URL.
  - **304 Not Modified:** O recurso não foi modificado desde a última requisição.

- **4xx: Erros do cliente**
  - **400 Bad Request:** A requisição não pode ser processada devido a um erro do cliente, como sintaxe de requisição inválida.
  - **401 Unauthorized:** O cliente não está autenticado para acessar o recurso solicitado.
  - **403 Forbidden:** O cliente está autenticado, mas não tem permissão para acessar o recurso.
  - **404 Not Found:** O recurso solicitado não foi encontrado no servidor.

- **5xx: Erros do servidor**
  - **500 Internal Server Error:** O servidor encontrou uma situação que não sabe como lidar.
  - **502 Bad Gateway:** O servidor, ao atuar como gateway ou proxy, recebeu uma resposta inválida do servidor upstream.
  - **503 Service Unavailable:** O servidor não está pronto para lidar com a requisição, geralmente devido a manutenção ou sobrecarga.

### **Como lidar com diferentes status codes em Python**

Ao usar a biblioteca `requests` em Python, você pode verificar o status code de uma resposta usando o atributo `status_code` do objeto de resposta. Aqui está um exemplo básico de como lidar com diferentes status codes:

```python
import requests

response = requests.get('https://api.exemplo.com/dados')

if response.status_code == 200:
    print("Requisição bem-sucedida!")
    data = response.json()
elif response.status_code == 404:
    print("Erro: Recurso não encontrado.")
elif response.status_code >= 500:
    print("Erro no servidor. Tente novamente mais tarde.")
else:
    print(f"Erro desconhecido. Status code: {response.status_code}")
```

É uma boa prática sempre verificar o status code ao fazer requisições para lidar adequadamente com possíveis erros e garantir que sua aplicação continue funcionando de forma suave.

### O que é JSON?



JSON (JavaScript Object Notation) é um formato de dados muito utilizado para transmitir dados estruturados pela web. Ele é baseado na sintaxe de objetos JavaScript, mas pode ser lido e interpretado por muitas linguagens de programação modernas, como Python.

O JSON é um formato de dados muito popular, pois é fácil de ler e escrever, além de ser leve e independente de linguagem. Ele é frequentemente usado para transmitir dados entre um servidor e um cliente da web.

Aqui está um exemplo de um objeto JSON:

```json
{
    "nome": "João",
    "idade": 30,
    "cidade": "São Paulo"
}
```

Observe que o JSON consiste em pares de chave-valor, onde as chaves são sempre strings e os valores podem ser strings, números, booleanos, objetos ou arrays. Os objetos JSON são delimitados por chaves `{}` e os arrays JSON são delimitados por colchetes `[]`.

Aqui está um exemplo de um array JSON:

```json
[
    {
        "nome": "João",
        "idade": 30,
        "cidade": "São Paulo"
    },
    {
        "nome": "Maria",
        "idade": 27,
        "cidade": "Rio de Janeiro"
    },
    {
        "nome": "Pedro",
        "idade": 35,
        "cidade": "Belo Horizonte"
    }
]
```

Qual é a diferença entre objetos e arrays JSON? Objetos JSON são delimitados por chaves `{}` e contêm pares de chave-valor. Arrays JSON são delimitados por colchetes `[]` e contêm uma lista de valores separados por vírgula. Objetos JSON são usados para representar registros individuais, enquanto arrays JSON são usados para representar uma lista de registros.

E qual é a diferença entre JSON e um dicionário do Python? JSON é um formato de dados independente de linguagem, enquanto um dicionário do Python é um tipo de dados específico da linguagem Python. JSON é uma string que pode ser lida e interpretada por muitas linguagens de programação modernas, incluindo Python. Dicionários do Python são estruturas de dados nativas da linguagem Python. JSON e dicionários do Python são semelhantes, mas não são a mesma coisa.
